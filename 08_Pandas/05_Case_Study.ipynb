# Pandas case study 
we will check the data from kashti(Titanic) dataset

```python
#Import libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
kashti=sns.load_dataset("titanic")
kashti.head(5)
```

### __Saving dataframe into csv file__

```python
#Saving dataframe into csv file
kashti.to_csv("kashti.csv")
```

### __To describe data__

```python
kashti.describe()
```

```python
kashti.head()
```

### __Dropping columns__

```python
#Dropping few columns and make new data set we will give axis which we want to remove means 1 for columns and 0 for rows
new_kashti=kashti.drop(["deck","embark_town","alone"],axis=1)
new_kashti.head()
```

```python
#Another way
kashti.drop(["deck","embark_town","alone"],axis=1).head()
```

### __Mean of data__

```python
kashti.mean()
```

### __Counting survived persons__

```python
kashti.value_counts("survived")
```

### __Grouping__

```python
#Grouping
g1=kashti.groupby(["sex","class"]).mean()
g1
```

```python
plt.figure(figsize=(15,10))
sns.set_theme(style="darkgrid",color_codes=True)
sns.barplot(x="survived",y="age",hue="fare",data=g1,palette="pastel")
plt.xlabel("Survived",size=15,weight="bold")
plt.ylabel("Age",size=15,weight="bold")
plt.title("Barplot of Titanic data grouped by their sex and class(Mean_data)",size=20)
plt.show()
```

### __Data of survived children__

```python
#Data of survived children
kashti[kashti["age"]<18].groupby(["sex","class"]).mean()
```
