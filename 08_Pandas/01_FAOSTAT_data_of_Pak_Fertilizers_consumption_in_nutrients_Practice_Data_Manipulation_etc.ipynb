     > ___We stored the data in p variable for plotting this data___

```python
p=df2.loc[ 1:30 , ["Value","Item"]]
p
```

> ___Barplot of variable p data___

```python
plt.figure(figsize=(12,10))
sns.set_theme(style="darkgrid",palette="deep")
sns.barplot(x="Value",y="Item",data=p,ci=None,saturation=1,order=("Nutrient nitrogen N (total)","Nutrient phosphate P2O5 (total)","Nutrient potash K2O (total)"))
plt.xlabel("Value(Tonnes)",size=13,weight="bold")
plt.ylabel("Item",size=13,weight="bold")
plt.title("Barplot of Pakistan Fertilizer consumption in nutrients",size=15,weight="bold")
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.boxplot(x="Value",y="Item",data=p,dodge=False,saturation=1,order=("Nutrient nitrogen N (total)","Nutrient phosphate P2O5 (total)","Nutrient potash K2O (total)"))
plt.xlabel("Value(Tonnes)",size=13,weight="bold")
plt.ylabel("Item",size=13,weight="bold")
plt.title("Boxplot of Pakistan Fertilizer consumption in nutrients",size=15,weight="bold")
plt.show()
```
