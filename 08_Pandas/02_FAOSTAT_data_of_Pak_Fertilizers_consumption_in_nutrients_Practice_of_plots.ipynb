# FAO Nutrients data consumption in Pakistan

> ___Importing libraries and reading data through csv___

```python
import pandas as pd
fao=pd.read_csv("FAOSTAT_data_of_Pak_Fertilizers_consumption_in_nutrients.csv")
fao
fao.head()
```

# Lineplot of nureients consumption in Pakistan
- Palettes for seaborn are:
  - dark
  - bright
  - pastel
  - colorblind
  - deep
  - muted

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(20,10))
sns.set_theme(style="ticks",color_codes=True)
sns.lineplot(x="Year", y="Value",data=fao,hue="Item",palette="dark" )
plt.xlabel("Year",size=20,weight="bold")
plt.ylabel("Value(Tonnes)",size=20,weight="bold")
plt.title("Pakistan Fertilizers consumption in nutrients",size=30,weight="bold")
# plt.savefig("Consumption of nutrients.png",dpi=300)
plt.show()
```

## Barplot of nutrients 

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(26,12))
sns.set_theme(style="ticks",color_codes=True)
sns.barplot(x="Item", y="Value",data=fao,hue="Year" ,saturation=0.8,order=("Nutrient nitrogen N (total)","Nutrient phosphate P2O5 (total)","Nutrient potash K2O (total)"))
plt.xlabel("Value(Tonnes)",size=20,weight="bold")
plt.ylabel("Item(Nutrient nitrogen N (total) & (Nutrient phosphate P2O5 (total))",size=20,weight="bold")
plt.title("Pakistan Fertilizers consumption in nutrients",size=30,weight="bold")
plt.savefig("Consumption of nutrients.png",dpi=300)
plt.show()
```

```python
df3=fao.loc[1:10,["Year","Item"]]
df3
```

```python
plt.figure(figsize=(20,15))
sns.violinplot(x="Item",y="Year",data=df3,saturation=1,order=("Nutrient nitrogen N (total)","Nutrient phosphate P2O5 (total)"))
plt.xlabel("Item(Nutrient nitrogen N (total) & (Nutrient phosphate P2O5 (total))", size=13,weight="bold")
plt.ylabel("Year",size=13,weight="bold")
plt.title("Violin plot of nuetrient consumption in pak",size=15,weight="bold")
plt.show()
```
