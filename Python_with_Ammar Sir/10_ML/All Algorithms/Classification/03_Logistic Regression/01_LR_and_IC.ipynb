```python
# Importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Import online data
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
# input variables ya features (X)
digits.data.shape
X=digits.data
# Means 1797 pictures size=64 8 by 8 (8*8)
```

```python
# output labels (Y)
digits.target.shape
y=digits.target
```

```python
# plotting 
plt.figure(figsize=(20,4))
# digits.data[0:10] means 10 images from data or input and digits.target[0:10] means labels or output
for index, (image , label) in enumerate (zip(digits.data[0:10] , digits.target[0:10])):
    plt.subplot(1 , 10 , index + 1) # index + 1 to end loop otherwise zero zero will keep running 
    plt.imshow(np.reshape(image , (8,8)) , cmap=plt.cm.gray) # (8,8) means 8 by 8 images equal to 64 our data set
    plt.title('Training: %i\n' % label , fontsize=20)
```

```python
# plotting 20 images
plt.figure(figsize=(20,4))
# digits.data[0:10] means 10 images from data or input and digits.target[0:10] means labels or output
for index, (image , label) in enumerate (zip(digits.data[0:20] , digits.target[0:20])):
    plt.subplot(1 , 20 , index + 1) # index + 1 to end loop otherwise zero zero will keep running 
    plt.imshow(np.reshape(image , (8,8)) , cmap=plt.cm.gray) # (8,8) means 8 by 8 images equal to 64 our data set
    plt.title( label , fontsize=20)
```

```python
# split the data
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X , y , test_size=1/5 , random_state=0)
```

```python
print("Train input data=",X_train.shape)
print("Test input data=",X_test.shape)
print("Train output data=",y_train.shape)
print("Test output data=",y_test.shape)
```

```python
# model train
from sklearn.linear_model import LogisticRegression
model = LogisticRegression().fit(X_train , y_train)
model
```

```python
model.predict(X_test[0:10])
prediction = model.predict(X_test)
prediction
```

```python
# Accuracy
# from sklearn.metrics import accuracy_score
score = model.score(X_test , y_test)
print("The accuracy score is = " , score)
```

```python
# Confusion matrix
from sklearn import metrics
cm= metrics.confusion_matrix(y_test , prediction)
cm
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()

# Draw a heatmap with the numeric values in each cell
f, ax = plt.subplots(figsize=(9, 6))
sns.heatmap(cm, annot=True, fmt="d", linewidths=.5, ax=ax)
plt.xlabel("Predicted value" , size=20)
plt.ylabel("Actutal value" , size=20)
all_sample_title= 'Accuracy Score : {0}'.format(score)
plt.title(all_sample_title , size=20)



# Means this plot tells us that our alogorithim is confused at points where it is showing 1 like on y-axis=1 and x-axis=8 it confused
```

```python
# Getting misclassified labels

import numpy as np 
import matplotlib.pyplot as plt
index = 0
misclassifiedIndexes = []
for label, predict in zip(y_test , prediction):
 if label != predict: 
  misclassifiedIndexes.append(index)
  index +=1
```

```python
plt.figure(figsize=(20,4))
for plotIndex, badIndex in enumerate(misclassifiedIndexes[0:5]):
 plt.subplot(1, 5, plotIndex + 1)
 plt.imshow(np.reshape(X_test[badIndex], (8,8)), cmap=plt.cm.gray)
 plt.title("Predicted: {}, Actual: {}".format(prediction[badIndex], y_test[badIndex]), fontsize = 15)
```
